```python
sentiment_mapping = {
    'Positive': 1,
    'Neutral': 0,
    'Negative': -1,
    'Neutral':2
}
df['sentiment_numeric'] = df['sentiment'].map(sentiment_mapping)
```

```python

max_positive_sentiment = df[df['sentiment_numeric'] == 1]


most_negative_sentiment = df[df['sentiment_numeric'] == -1]
```

```python

print("Maximum Positive Sentiment:")
print(max_positive_sentiment)


print("Most Negative Sentiment:")
print(most_negative_sentiment)
```

```python

max_positive_sentiment = df[df['sentiment'] == df['sentiment'].max()]
most_negative_sentiment = df[df['sentiment'] == df['sentiment'].min()]


print("Maximum Positive Sentiment:")
print(max_positive_sentiment)

print("Most Negative Sentiment:")
print(most_negative_sentiment)
```

```python

topic_sentiment = df.groupby('topic')['sentiment_numeric'].mean()


most_positive_topic = topic_sentiment.idxmax()
most_positive_score = topic_sentiment.max()

print(f"Topic with the most positive sentiment: {most_positive_topic}")
print(f"Average sentiment score: {most_positive_score}")
```

```python

topic_sentiment = df.groupby('topic')['sentiment_numeric'].mean()


most_negative_topic = topic_sentiment.idxmin()
most_negative_score = topic_sentiment.min()

print(f"Topic with the most negative sentiment: {most_negative_topic}")
print(f"Average sentiment score: {most_negative_score}")
```

```python

sentiment_counts = df.groupby('topic').size()
neutral_counts = df[df['sentiment'] == 'Neutral'].groupby('topic').size()


neutral_proportion = (neutral_counts / sentiment_counts).fillna(0)


most_neutral_topic = neutral_proportion.idxmax()
most_neutral_proportion = neutral_proportion.max()

print(f"Topic with the most neutral sentiment: {most_neutral_topic}")
print(f"Proportion of neutral sentiment: {most_neutral_proportion:.2f}")

```

```python

# Create a crosstab to count occurrences of each combination of Topic and Sentiment
crosstab = pd.crosstab(index=df['topic'], columns=df['sentiment'])

# Set up the plot size
plt.figure(figsize=(12, 8))

# Create the heatmap
sns.heatmap(
    crosstab,                 # Data for the heatmap
    cmap='coolwarm',          # Color map
    annot=True,               # Annotate each cell with the numeric value
    fmt='d',                  # Format for annotation (decimal integer)
    linewidths=.5             # Width of lines that will divide each cell
)

# Add title and labels
plt.title('Heatmap of Topic vs Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Topic')

# Display the plot
plt.show()
```
